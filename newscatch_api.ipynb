```python
# Default packages
import time
import csv
import os
import json
import requests
import json
import pandas
import dotenv
```

```python
env_var = dotenv.dotenv_values(dotenv_path='.env')
```

```python
# URL of our NewsCatcher API
base_url = 'https://api.newscatcherapi.com/v2/search'

api_key = env_var['NEWS_API_KEY']
```

```python
# Put your API key to headers in order to be authorized to perform a call
headers = {'x-api-key': api_key}

# Define your desired parameters
params = {
    'q': 'Bitcoin AND Ethereum AND Dogecoin',
    'lang': 'en',
    'to_rank': 10000,
    'page_size': 100,
    'page': 1
    }

# Make a simple call with both headers and params
response = requests.get(base_url, headers=headers, params=params)

# Encode received results
results = json.loads(response.text.encode())
if response.status_code == 200:
    print('Done')
else:
    print(results)
    print('ERROR: API call failed.')
```
